```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Initialise x0 as value at t0
x0 = 1

#Initialises array x and t with 0
x= np.zeros(100) #for forward euler method
y= np.zeros(100) #for backward euler method
t= np.linspace(0, 7, num=100)
delt= 0.07

#Assigning initial value
x[0]=x0
y[0]=x0
```

```python
#Set up a loop to find subsequent values of x at each point
#FORWARD EULER
for i in range(99):
    x[i+1]= x[i] + (delt*2*x[i])    
```

```python
#Real function would be
#REAL FUNCTION
real_x= np.exp(2*t)
```

```python
#Set up a loop to find subsequent values of x at each point
#BACKWARD EULER
for i in range(99):
    y[i+1]= y[i]/(1- (delt*2))   
```

```python
plt.figure(figsize=(12,5))
plt.plot(t, x, 'b-', label='Forward Euler method')
plt.plot(t, y, 'g-', label='Backward Euler method')
plt.plot(t, real_x, 'r-', label= 'Real curve')

plt.xlabel('Time')
plt.ylabel('X')
plt.legend(loc='best')
plt.title('Euler Methods (dx/dt = 2x)')
plt.show()
```
